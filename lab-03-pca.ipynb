```python
import pandas as pd

# Data is Wisconsin Diagnosis Breast Cancer -- from https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+(Diagnostic).
# Load the dataset and have a look. (Names for the dataset are in another file, annoyingly.)
df = pd.read_csv('wdbc.data', header=None)
df.head()
```

```python
# The outcome (target) = 'M' (malignant) or 'B' (benign).
# It is in the column at index 1
df.iloc[:,1].unique()
```

```python
# Feature labels are as follows:
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error', 'perimeter error', 'area error',
       'smoothness error', 'compactness error', 'concavity error',
       'concave points error', 'symmetry error',
       'fractal dimension error', 'worst radius', 'worst texture',
       'worst perimeter', 'worst area', 'worst smoothness',
       'worst compactness', 'worst concavity', 'worst concave points',
       'worst symmetry', 'worst fractal dimension']
```

```python
df.columns = column_labels
df
# This could also be done with: df.rename(columns={0:  'patient ID', etc.
```

```python
# We may want to change the 'M' and 'B' labels to full names so it's easier to read. Or not.
df['diagnosis'].replace('B', 'Benign', inplace=True)
df['diagnosis'].replace('M', 'Malignant', inplace=True)
df
```

```python
# Put the features, labels into their own dataframes
feature_df = df.iloc[:,2:]
label_df = df.iloc[:,1]
```

```python
feature_df
```

```python
label_df
```

```python
# Use sklearn to standardise (specifically: normalise) the data so that 
# the data has a normal distribution -- mean = 0; stdev = 1
from sklearn.preprocessing import StandardScaler
import numpy as np

x = feature_df.values
x = StandardScaler().fit_transform(x)
# ... and check if it worked
np.mean(x),np.std(x)
```

```python
# x is now a numpy array. Convert it back to a df?
# ******************************
# WHY?!?!?
# ******************************
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.tail()
normalised_breast
```

```python
# Choose the first 2 principal components
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)

principal_breast_df = pd.DataFrame(data = principalComponents_breast
             , columns = ['principal component 1', 'principal component 2'])
```

```python
# One thing that's often useful is "how much of the variance is explained"
# in any principal component. We can find out that 44.28% is explained
# by PC1; 18.97 by PC2:
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
# Let's plot it!
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component 1',fontsize=20)
plt.ylabel('Principal Component 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['b', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_breast_df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

```python
# How many principal components do we need to explain the variance in data?
pca = PCA().fit(x)

evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))
# FYI: We have to go to 6 significant digits to explain the last PC.
```

```python
# Graph it...
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance');
```

```python
import pandas as pd

# Load the dataset and have a look. (Names for the dataset are in another file, annoyingly.)
df = pd.read_csv('gisette_train.data', header=None, delim_whitespace = True)
# df.drop(columns=5000)
# print(df)
df

#read data and deliminate the NaN values
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

x = df.values
x = StandardScaler().fit_transform(x)
# ... and check if it worked
np.mean(x),np.std(x)

#fit(): Method calculates the parameters Î¼ and Ïƒ and saves them as internal objects.
#transform(): Method using these calculated parameters apply the transformation to 
#             a particular dataset.
#fit_transform = fit()+transform()
```

```python
from sklearn.decomposition import PCA

pca_gisette = PCA(n_components=2) 
principalComponents_gisette = pca_gisette.fit_transform(x)
principal_gisette_df = pd.DataFrame(data = principalComponents_gisette
             , columns = ['principal component 1', 'principal component 2'])
principal_gisette_df

#new a PCA object pca_gisette, set PC = 2, because we want to draw 2D picture
#since there are 5000 features, there are 5000 pc's, 
#                              in which pc1 and pc2 are the most relavent pc
#put x's information into principalComponents_gisette
#make a dataframe with pc1 and pc2
```

```python
principal_gisette_df.sum(axis = 0)
```

```python
print('Explained variation per principal component: {}'
      .format(pca_gisette.explained_variance_ratio_))
#we can see pc1 and pc2 count 0.029 and 0.023 respectively, in 5000 pc's
#the sum of all pc = 1, so 0.029 and 0.023 are the greater part
```

```python
label_df = pd.read_csv('gisette_train.labels', header=None, delim_whitespace = True)
label_df = label_df.iloc[:,0]

# gisette_train.labels
#this label contains 1 and -1.
#i guess it's the classification of 4 and 9
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component 1',fontsize=20)
plt.ylabel('Principal Component 2',fontsize=20)
plt.title("Principal Component Analysis of Gisette Dataset",fontsize=20)
targets = [1, -1]
colors = ['b', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = label_df == target
    plt.scatter(principal_gisette_df.loc[indicesToKeep, 'principal component 1']
               , principal_gisette_df.loc[indicesToKeep, 'principal component 2'], c = color, s =2)

plt.legend(targets,prop={'size': 15})

#drawing
#the plot shows there are more cases "1" when pc1<-10
#but for pc1>-10, the samples are almost overlapped
```

```python
pca = PCA().fit(x)

evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}: {:0.4f}'.format(*k) for k in enumerate(evr)))

#it shows every principal component
#we can see pc1 = 0.0288, pc2 = 0.0226, and after pc1797, they only contribute 0.0001
#after pc2844, they are zero, which means not important features
```

```python
# Graph it...
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance')

#this is cumulated sum.
#we can see the sum of all pc is "1"
```
